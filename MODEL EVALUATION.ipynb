# MODEL  BUILDING

```python
#Importing basic libraries
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
#Loading dataset
df=pd.read_csv('cleanchurn.csv',index_col=0)
df
```

```python
#Divide data into X and Y
X=df.iloc[:,0:18]
Y=df.iloc[:,18]
```

```python
X
```

```python
Y
```

# SCALING

```python
#Importing Scaling Libraries
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler
```

```python
ls=X.columns
```

```python
ls=ls.values
```

```python
ls=np.delete(ls,[1,3])
```

```python
scaler1 =MinMaxScaler().fit(X[ls])
#scaler2 =StandardScaler().fit(X[ls])
#scaler3 =RobustScaler().fit(X[ls])
X1=X
#X2=X
#X3=X
X1[ls]=scaler1.transform(X[ls])
#X2[ls]=scaler2.transform(X[ls])
#X3[ls]=scaler3.transform(X[ls])
Y1=Y.values
```

```python
#Importing Train test split lib
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=5)
x1train,x1test,y1train,y1test=train_test_split(X1,Y1,test_size=0.2,random_state=5)
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
#Importing CV searched libraries
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV,StratifiedKFold
```

```python
cv=StratifiedKFold(n_splits=10,shuffle=True,random_state=24)
```

```python
#Importing Model Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import  DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Importing report libraries
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
#Importing Smote 
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=24)
```

```python
X_train_smote, y_train_smote = smote.fit_resample(xtrain.astype('float'),ytrain)
```

```python
y_train_smote.value_counts()
```

#  1 Logistic Regression

```python
model1=LogisticRegression()

# Cross validation score
scores=cross_val_score(model1,xtrain,ytrain,cv=10)
print('Average Cross val score',scores.mean())

# Accuracy Score
model1.fit(xtrain,ytrain)
pred=model1.predict(xtest)
predt=model1.predict(xtrain)
print("Train Model accuracy",accuracy_score(predt,ytrain))
print('Test Model accuracy',accuracy_score(pred,ytest))

# Confusion Matrix
print(confusion_matrix(pred,ytest))

# classification report
print(classification_report(pred,ytest))

#Error
print("Mean Squared Error:",mean_squared_error(pred,ytest))
print("Mean Absolute Error:",mean_absolute_error(pred,ytest))
print("Mean Absolute percentage Error:",mean_absolute_percentage_error(pred,ytest))
```

#  Logistic Regression Using Smote dataset

```python
model1=LogisticRegression()

# Cross validation score
scores=cross_val_score(model1,X_train_smote,y_train_smote,cv=cv)
print('Average Cross val score',scores.mean())

# Accuracy Score
model1.fit(X_train_smote,y_train_smote)
pred=model1.predict(xtest)
predt=model1.predict(X_train_smote)
print("Train Model accuracy",accuracy_score(predt,y_train_smote))
print('Test Model accuracy',accuracy_score(pred,ytest))

# Confusion Matrix
print(confusion_matrix(pred,ytest))

# classification report
print(classification_report(pred,ytest))

#Error
print("Mean Squared Error:",mean_squared_error(pred,ytest))
print("Mean Absolute Error:",mean_absolute_error(pred,ytest))
print("Mean Absolute percentage Error:",mean_absolute_percentage_error(pred,ytest))
```
